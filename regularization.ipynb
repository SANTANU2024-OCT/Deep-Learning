```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X, y = make_moons(100, noise=0.25,random_state=2)
```

```python
X
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='Set1')
plt.title('Visualization of make_moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model1=Sequential()
model1.add(Dense(128,input_dim=2,activation='relu'))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
from tensorflow.keras.optimizers import Adam
adam=Adam(learning_rate=0.001)
model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
history1 = model1.fit(X, y, epochs=2000, validation_split = 0.2,verbose=1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(history1.history['loss'], label='Training Loss')
plt.plot(history1.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the boundaries for the plot
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

# Create a meshgrid to plot the decision boundary
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict probabilities for each point in the meshgrid
Z = model1.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm', levels=np.linspace(0, 1, 100))
plt.colorbar(label='Predicted Probability')

# Overlay the original data points
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='Set1', s=80, edgecolor='k')

plt.title('Neural Network Decision Region for make_moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
model2 = Sequential()

model2.add(Dense(128,input_dim=2, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l2(0.0001)))
model2.add(Dropout(0.2))
model2.add(Dense(128, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l2(0.001)))
model2.add(Dropout(0.2))
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
from tensorflow.keras.optimizers import Adam
adam2=Adam(learning_rate=0.001)
model2.compile(loss='binary_crossentropy', optimizer=adam2, metrics=['accuracy'])
```

```python
history2 = model2.fit(X, y, epochs=2000, validation_split = 0.2,verbose=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(history2.history['loss'], label='Training Loss')
plt.plot(history2.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
